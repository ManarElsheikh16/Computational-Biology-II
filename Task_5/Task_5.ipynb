```python
import networkx as nx
import numpy as np
```

```python
graph=nx.Graph()
```

```python
graph.add_nodes_from([1,2,3,4])
graph.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
```

```python
nx.draw(graph,with_labels=True,node_color='orange',node_size=2500,font_size=35,font_color='white')
```

```python
dict(graph.degree)
```

```python
nx.to_numpy_matrix(graph)
```

```python
import pandas as pd
```

```python
df=pd.read_csv("HomoSapiens_binary_hq.txt",nrows=100,sep='\t',usecols=['Gene_A','Gene_B'])
```

```python
df
```

```python
Genes=nx.from_pandas_edgelist(df,'Gene_A','Gene_B',create_using=nx.DiGraph())        #source and destination  'Gene_A','Gene_B'
```

```python
from scipy.sparse.csgraph import connected_components
```

```python
#Biological Network
nx.draw(Genes,with_labels=True,node_size=1000,node_color="purple",font_color="orange")
```

```python
import matplotlib.pyplot as plt
```

```python
degrees=dict(Genes.degree())
```

```python
order=sorted(degrees.items(),key=lambda t :t[1])
```

```python
x,y=zip(*order)
```

```python
plt.plot(x,y)
plt.show()
```
